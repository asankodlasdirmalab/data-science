## TapŇüńĪrńĪq 1 
AŇüańüńĪdakńĪ arrayd…ô birinci elementi, sonuncu elementi v…ô 3-c√ľ indeksd…ôki elementi ekrana g√∂st…ôrin

`arr = np.array([12, 25, 37, 48, 59, 63])`

````python
import numpy as np

arr = np.array([12, 25, 37, 48, 59, 63])

print("Birinci element:", arr[0])
print("Sonuncu element:", arr[-1])
print("3-c√ľ indeksd…ôki element:", arr[3])
````

## TapŇüńĪrńĪq 2
AŇüańüńĪdakńĪ arrayda ilk 4 elementi, son 3 elementi v…ô h…ôr ikinci elementi slicing edin v…ô ekranda g√∂st…ôrin

`arr = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45])`

````python
arr = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45])

print("ńįlk 4 element:", arr[:4])
print("Son 3 element:", arr[-3:])
print("H…ôr ikinci element:", arr[::2])
````

## TapŇüńĪrńĪq 4 
AŇüańüńĪdakńĪ matrix √ľ√ß√ľn birinci s√ľtunu, sonuncu s√ľtunu v…ô sonuncu s…ôtri slicing il…ô ekranda g√∂st…ôrin

## TapŇüńĪrńĪq 5 
AŇüańüńĪdakńĪ arrayd…ô 30-dan b√∂y√ľk olan elementl…ôri, c√ľt …ôd…ôdl…ôri v…ô 50-y…ô b…ôrab…ôr olan elementl…ôri boolean indexing metodu il…ô …ôld…ô edin

`arr = np.array([10, 15, 20, 35, 40, 50, 65])`

````python
arr = np.array([10, 15, 20, 35, 40, 50, 65])

print("30-dan b√∂y√ľk elementl…ôr:", arr[arr > 30])
print("C√ľt …ôd…ôdl…ôr:", arr[arr % 2 == 0])
print("50-y…ô b…ôrab…ôr olan elementl…ôr:", arr[arr == 50])
````

## TapŇüńĪrńĪq 6 ‚ÄĒ np.where()
AŇüańüńĪdakńĪ arrayd…ô 28 elementinin indeksini v…ô 20-d…ôn b√∂y√ľk olan elementl…ôrin indeksl…ôrini tapńĪn

`arr = np.array([7, 14, 21, 28, 35, 42])`

````python
arr = np.array([7, 14, 21, 28, 35, 42])

print("28 elementinin indeksi:", np.where(arr == 28)[0])
print("20-d…ôn b√∂y√ľk olan elementl…ôrin indeksl…ôri:", np.where(arr > 20)[0])
````

## TapŇüńĪrńĪq 7 
AŇüańüńĪdakńĪ m…ôlumatlar √ľ√ß√ľn b√ľt√ľn t…ôl…ôb…ôl…ôrin adlarńĪnńĪ (slicing), b√ľt√ľn ballarńĪ (slicing) v…ô 80-d…ôn y√ľks…ôk bal toplayan (boolean indexing) t…ôl…ôb…ôl…ôri …ôld…ô edin v…ô ekranda g√∂st…ôrin

```python
students = np.array([
    ["Ali", 75],
    ["Aysel", 92],
    ["Murad", 68],
    ["Leyla", 88],
    ["Nigar", 95]
], dtype=object)
```

````python
students = np.array([
    ["Ali", 75],
    ["Aysel", 92],
    ["Murad", 68],
    ["Leyla", 88],
    ["Nigar", 95]
], dtype=object)

names = students[:, 0]
grades = students[:, 1]

print("B√ľt√ľn t…ôl…ôb…ôl…ôrin adlarńĪ:", names)
print("B√ľt√ľn ballar:", grades)
print("80-d…ôn y√ľks…ôk bal toplayan t…ôl…ôb…ôl…ôr:\n", students[grades > 80])
````

## TapŇüńĪrńĪq 8 
AŇüańüńĪdakńĪ m…ôlumatlar √ľ√ß√ľn m…ôhsul adlarńĪnńĪ, qiym…ôtl…ôri v…ô qiym…ôti 100 manatdan √ßox olan m…ôhsullarńĪ …ôld…ô edin

```python
products = np.array([
    ["Mouse", 35],
    ["Keyboard", 120],
    ["Monitor", 450],
    ["USB", 15],
    ["Headset", 180]
], dtype=object)
```

````python
products = np.array([
    ["Mouse", 35],
    ["Keyboard", 120],
    ["Monitor", 450],
    ["USB", 15],
    ["Headset", 180]
], dtype=object)

prod_names = products[:, 0]
prices = products[:, 1]

print("M…ôhsul adlarńĪ:", prod_names)
print("Qiym…ôtl…ôr:", prices)
print("Qiym…ôti 100 manatdan √ßox olan m…ôhsullar:\n", products[prices > 100])
````

## TapŇüńĪrńĪq 9 ‚ÄĒ Yekun
AŇüańüńĪdakńĪ matrix √ľ√ß√ľn:
1. Birinci s…ôtri g√∂st…ôrin
2. Sonuncu s√ľtunu g√∂st…ôrin
3. ńįlk iki s…ôtri g√∂st…ôrin
4. 50-d…ôn b√∂y√ľk olan elementl…ôri √ßńĪxarńĪn
5. 70 elementinin indeksini tapńĪn

```python
arr = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])
```

````python
arr = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print("1. Birinci s…ôtir:", arr[0, :])
print("2. Sonuncu s√ľtun:", arr[:, -1])
print("3. ńįlk iki s…ôtir:\n", arr[:2, :])
print("4. 50-d…ôn b√∂y√ľk olan elementl…ôr:", arr[arr > 50])

coords = np.where(arr == 70)
print("5. 70 elementinin indeksi (s…ôtir, s√ľtun):", (coords[0][0], coords[1][0]))
````
